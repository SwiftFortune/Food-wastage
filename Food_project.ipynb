<p style = "color:pink;, Front-size: 20px;">18. How many food claims have been made for each food item?<p/>

```python
query18 = """ 
    SELECT Status, COUNT(*) AS Delivered
    FROM petitions
    WHERE Status = 'Completed' 
    GROUP BY Status
    ORDER BY Delivered ASC;
"""
cursor.execute(query18)
result = cursor.fetchall()
df = pd.DataFrame(result, columns = ["Status", "Delivered"])
df
```

<p style="color:pink;, Front-size:20px;">19. What percentage of food claims are completed vs. pending vs. canceled?<p/>

```python
query19 = """ 
   SELECT Status, COUNT(*) AS Total_Claims,
    ROUND((COUNT(*) * 100.0 / (SELECT COUNT(*) FROM petitions)), 2) AS Percentage
    FROM petitions
    GROUP BY Status;
""" 
cursor.execute(query19)
result = cursor.fetchall()
df = pd.DataFrame(result, columns = ["Status","Total_Claims","Percentage"])
df
```

<p style = "color:pink;, Front-size:20px;">20. How many are there in pending?<p/>

<p style = "color:pink;, Front-size: 20px;">20. Find expiring food Items that have not been claimed yet?<p/>

```python
query20 = """ 
    SELECT f.Food_ID, f.Food_Name, f.Quantity, f.Expiry_Date, f.Location 
    FROM provisions AS f
    LEFT JOIN petitions AS c ON f.Food_ID = c.Food_ID
    WHERE c.Food_ID IS NULL
    AND f.Expiry_Date <= CURDATE()
    ORDER BY f.Expiry_Date ASC;
    """
cursor.execute(query20)
result = cursor.fetchall()
df = pd.DataFrame(result, columns=["Food_ID", "Food_Name", "Quantity", "Expiry_Date", "Location"])
df
```

<p style = "color:pink;, Front-size: 20px;">21.Find Food Claims with Full Details<p/>

```python
query21 = """ 
    SELECT a.Claim_ID, a.Food_ID, a.Receiver_ID, a.Status, a.Timestamp,
    b.Food_Name, b.Quantity, b.Expiry_Date, b.Provider_Type, b.Location, b.Food_type, b.Meal_Type
    FROM petitions AS a
    INNER JOIN provisions AS b
    ON a.Food_ID = b.Food_ID
    """
cursor.execute(query21)
result = cursor.fetchall()
df = pd.DataFrame(result, columns =["Claim_ID","Food_ID","Receiver_ID","Status","Timestamp","Food_Name","Quantity","Expiry_Date","Provider_Type","Location","Food_type","Meal_Type"])
df
```

<p style = "color:pink;, Front-size:20px;">22.List unclaimed food listing by Restaurant<p/>

<p style = "color:pink;, Front-size:20px;">24. What is the average quantity of food claimed per receiver?<p/>

```python
query24 = """ 
    SELECT Receiver_ID, AVG(Quantity) FROM petitions
    JOIN provisions ON petitions.Food_ID = provisions.Food_ID
    GROUP BY Receiver_ID
    ORDER BY AVG(Quantity)
    """
cursor.execute(query24)
result = cursor.fetchall()
df = pd.DataFrame(result, columns = [i[0] for i in cursor.description])
df
```
